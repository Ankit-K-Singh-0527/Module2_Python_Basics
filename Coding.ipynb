```python
#10.  Write code to demonstrate type casting with list elements.


list=[0,8.8,True,9,'982',0b0110110]
print('org_list : ',list)
# using list comprehension!
print('int_list : ',[int(i) for i in list if isinstance(i,(int,float,bool)) or i.isdigit() ])
print('float_list : ',[float(i) for i in list if isinstance(i,(int,float,bool)) or i.isdigit()])
print('bool_list : ',[bool(i) for i in list if (isinstance(i,(int,float,bool) or i.isdigit()))])
```
